```python
import h5py
import os
from pathlib import Path

def count_total_samples(data_dir, num_frames=16, stride=1):
    """
    Calculate total number of training samples from raw HDF5 files.
    
    Args:
        data_dir: Path to directory containing HDF5 files
        num_frames: Frames per sample (context + target combined)
        stride: Sliding window stride
    
    Returns:
        Total samples and breakdown per file
    """
    hdf5_files = list(Path(data_dir).glob("*.hdf5")) + list(Path(data_dir).glob("*.h5"))
    
    if not hdf5_files:
        print(f"No HDF5 files found in {data_dir}")
        return
    
    total_samples = 0
    samples_per_file = {}
    
    for file_path in hdf5_files:
        try:
            with h5py.File(file_path, "r") as f:
                # Get number of timesteps (try different possible keys)
                if 'dimensions/time' in f:
                    num_timesteps = f['dimensions/time'].shape[0]
                elif 'time' in f:
                    num_timesteps = f['time'].shape[0]
                else:
                    # Fallback: check first field's time dimension
                    first_field = None
                    for key in f.keys():
                        if isinstance(f[key], h5py.Group):
                            for subkey in f[key].keys():
                                if isinstance(f[key][subkey], h5py.Dataset):
                                    first_field = f[key][subkey]
                                    break
                    
                    if first_field is None:
                        print(f"Could not determine timesteps in {file_path.name}")
                        continue
                    
                    num_timesteps = first_field.shape[1] if len(first_field.shape) > 1 else first_field.shape[0]
                
                # Calculate samples from this file
                # sliding window: [t0:t0+num_frames), [t0+stride:t0+stride+num_frames), ...
                num_samples = max(0, (num_timesteps - num_frames) // stride + 1)
                
                samples_per_file[file_path.name] = {
                    'timesteps': num_timesteps,
                    'samples': num_samples
                }
                total_samples += num_samples
                
        except Exception as e:
            print(f"Error processing {file_path.name}: {e}")
    
    # Print results
    print(f"\n{'File':<60} | {'Timesteps':<12} | {'Samples':<10}")
    print("=" * 85)
    for filename, info in sorted(samples_per_file.items()):
        print(f"{filename:<60} | {info['timesteps']:<12} | {info['samples']:<10}")
    
    print("=" * 85)
    print(f"Total files: {len(samples_per_file)}")
    print(f"Total samples: {total_samples}")
    print(f"(Config: num_frames={num_frames}, stride={stride})")
    
    return total_samples, samples_per_file


# Usage
if __name__ == "__main__":
    # Adjust these paths
    train_dir = "/home/skanda/active_matter/data/train"
    val_dir = "/home/skanda/active_matter/data/valid"
    test_dir = "/home/skanda/active_matter/data/test"
    
    print("TRAINING SET")
    train_total, _ = count_total_samples(train_dir, num_frames=8, stride=1)
    
    print("\n\nVALIDATION SET")
    val_total, _ = count_total_samples(val_dir, num_frames=8, stride=1)
    
    print("\n\nTEST SET")
    test_total, _ = count_total_samples(test_dir, num_frames=8, stride=1)
    
    print("\n\nSUMMARY")
    print(f"Train: {train_total}")
    print(f"Val: {val_total}")
    print(f"Test: {test_total}")
    print(f"Total: {train_total + val_total + test_total}")
```

```python
import h5py
import os
from pathlib import Path

def count_total_samples(data_dir, num_frames=16, stride=1):
    """Calculate total samples from raw HDF5 files."""
    hdf5_files = list(Path(data_dir).glob("*.hdf5")) + list(Path(data_dir).glob("*.h5"))
    
    if not hdf5_files:
        print(f"No HDF5 files found in {data_dir}")
        return 0, {}
    
    total_samples = 0
    samples_per_file = {}
    
    for file_path in sorted(hdf5_files):
        try:
            with h5py.File(file_path, "r") as f:
                # Get timesteps
                if 'dimensions/time' in f:
                    num_timesteps = f['dimensions/time'].shape[0]
                else:
                    continue
                
                # Calculate samples: (T - F) / stride + 1
                num_samples = max(0, (num_timesteps - num_frames) // stride + 1)
                
                samples_per_file[file_path.name] = {
                    'timesteps': num_timesteps,
                    'samples': num_samples
                }
                total_samples += num_samples
                
        except Exception as e:
            print(f"Error: {file_path.name}: {e}")
    
    return total_samples, samples_per_file

def print_results(split_name, total_samples, samples_per_file):
    """Print formatted results."""
    print(f"\n{'='*85}")
    print(f"{split_name.upper()}")
    print(f"{'='*85}")
    print(f"{'File':<60} | {'T':<4} | {'Samples':<8}")
    print("-"*85)
    
    for filename in sorted(samples_per_file.keys()):
        info = samples_per_file[filename]
        print(f"{filename:<60} | {info['timesteps']:<4} | {info['samples']:<8}")
    
    print("-"*85)
    print(f"Total files: {len(samples_per_file)}")
    print(f"Total samples: {total_samples}")

if __name__ == "__main__":
    train_dir = "/home/skanda/active_matter/data/train"
    val_dir = "/home/skanda/active_matter/data/valid"
    test_dir = "/home/skanda/active_matter/data/test"
    
    num_frames = 16
    
    print(f"\n\n{'#'*85}")
    print(f"CALCULATING WITH num_frames={num_frames}")
    print(f"{'#'*85}")
    
    train_total, train_files = count_total_samples(train_dir, num_frames=num_frames, stride=1)
    val_total, val_files = count_total_samples(val_dir, num_frames=num_frames, stride=1)
    test_total, test_files = count_total_samples(test_dir, num_frames=num_frames, stride=1)
    
    print_results("Training", train_total, train_files)
    print_results("Validation", val_total, val_files)
    print_results("Test", test_total, test_files)
    
    print(f"\n{'='*85}")
    print("FINAL SUMMARY")
    print(f"{'='*85}")
    print(f"Train:   {train_total:>8} samples")
    print(f"Val:     {val_total:>8} samples")
    print(f"Test:    {test_total:>8} samples")
    print(f"Total:   {train_total + val_total + test_total:>8} samples")
    print(f"\nExpected: 8750 train, 1200 val, 1300 test = 11250 total")
    print(f"Match: {'✓' if abs(train_total - 8750) < 100 else '✗'}")
```

```python
import os
import h5py
from glob import glob

train_dir = "/home/skanda/active_matter/data/train"
val_dir = "/home/skanda/active_matter/data/valid"
test_dir = "/home/skanda/active_matter/data/test"

from pathlib import Path
import h5py

F = 16
stride = 1

def count_samples(split_dir):
    paths = list(Path(split_dir).rglob("*.h5")) + list(Path(split_dir).rglob("*.hdf5"))

    print("Found files:", len(paths))

    total = 0

    for path in paths:
        with h5py.File(path, "r") as f:
            N, T, H, W = f["t0_fields/concentration"].shape
            total += N * max(0, (T - 2*F)//stride + 1)

    print("Total samples:", total)
    return total

count_samples(train_dir)
count_samples(val_dir)
count_samples(test_dir)
```
